# Length vs (P2 - P1)

```python
# del_p = p2 - p1 

q = 4  # flow rate
vis = 2  # viscosity
a = 500*30  # cross sectional area
k = 50  # permeability 
l = 2500 # length
p1 = 2100 # initial pressure

n = int(input("Enter the step size of length division: "))

import matplotlib.pyplot as plt 

l1 = [] # values of del p 
l2 = [] # values of length 

for i in range(0,2501,n):
  del_p = -(q*vis*i)/(0.001127*k*a)
  l1.append(del_p)
  l2.append(i)

plt.plot(l2,l1)
plt.xlabel('Length L (ft)')
plt.ylabel('Pressure difference (p2 - p1) psi')
plt.show()
```

# Length vs P2

P1 = 2100 psi

```python
q = 4  # flow rate
vis = 2  # viscosity
a = 500*30  # cross sectional area
k = 50  # permeability 
l = 2500 # length
p1 = 2100 # initial pressure

n = int(input("Enter the step size of length division: "))

import matplotlib.pyplot as plt 

l1 = [] # values of del p 
l2 = [] # values of length 

for i in range(0,2501,n):
  p2 = -((q*vis*i)/(0.001127*k*a)) + p1
  l1.append(p2)
  l2.append(i)

plt.plot(l2,l1)
plt.xlabel('Length L (ft)')
plt.ylabel('Final pressure P2 (psi)')
plt.show()
```
